```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = [1,2,3,10,5,10,7]
b = [10,2,35,24,66,1,7]
ax = plt.axes()
ax.set_xlim([0,70])
ax.set_ylim([0,10])
ax.set_xticks([i for i in range(0,71,10)])
ax.set_yticks([i for i in range(0,11,1)])
ax.grid()
plt.plot(b,a, color="r")
plt.plot(range(1,len(a)+1),a,color="black")
plt.xlabel("Roll No")
plt.ylabel("CPI")
plt.annotate(text="Lowest cpi",xy=[5,5])
plt.title("Exam Review")
plt.legend(["CX","CY"],loc = 4)
plt.show()
```

```python
y = [85,70,60,82];
x = ["C++","Java","Python","Dart"]
c = ["r","b","g","y"]
w = [0.4,0.5,0.6,0.7]
plt.bar(x,y,color=c,width=w,alpha=0.5,edgecolor='black')
plt.show()
```

```python
# pie chart
x = [300,400,500,200,100]
y = ["Food","Travel","Accomodation","Misc","Shopping"]
c= ["c","r","m","y","g"]
e= [0,0,0.1,0,0]
plt.pie(x,labels=y,colors=c,explode=e,autopct="%1.1f%%",shadow=True)
plt.show()
```

```python
y = [85,70,60,82];
x = ["C++","Java","Python","Dart"]
c = ["r","b","g","y"]
w = [0.4,0.5,0.6,0.7]
plt.barh(x,y,color=c,height=w,alpha=0.5,edgecolor='black')
plt.show()
```
