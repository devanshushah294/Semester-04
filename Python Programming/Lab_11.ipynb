### 01) WAP to demonstrate the use of Pie chart.

```python
x = [i for i in range(10,50,10)]
ex = [0.2,0.0,0.0,0.0]
y = ["food","travel","accomodation","shopping"]
plt.pie(
        x,labels=y,colors=['r','y','g','m'],
        radius=1,
        shadow=2,
        autopct="%0.1f%%",
        explode=ex,
        labeldistance=1.3,
        textprops={"fontsize":15,"color":"g"}
        );
# plt.pie(x);
plt.title("expenses",c="b")
plt.show();
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
def func_percentage(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i]);
x = ["python","c","c++","java"]
c = ['c','y','m','r'];
y = [85,70,60,82];
plt.bar(x,y,align="edge",color=c,edgecolor="b",alpha=0.5);
plt.xlabel("languages");
plt.ylabel("popularity in percentage");
plt.title("popularity of languages");
func_percentage(x,y);

plt.show();
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?
